```python
#factorial calculation
try:
    num=int(input("Please put in a non-negative integer:"))
    num_input=num
    if num<0:
        print("The number must be positive.")
    else:
        while num_input>0:
            print("{}".format(num_input))
            num-=1
except:
    print("Please re-enter a non-negative integer:")

```
